## Builld a model to predict the salary for an emloyee based on their position in organization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('C:/Users/ADMIN/Desktop/FBS/Data Science/Data sets/Position_Salaries.csv')
```

```python
dataset
```

```python
plt.boxplot(dataset['Salary'])
```

```python
plt.scatter(x=dataset['Level'],y = dataset['Salary'])
```

```python
X = dataset[['Level']]
y = dataset['Salary']
```

```python
## Linear regression implementation without polynomil conversion
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
predicted_data = lr.predict(X)
```

```python
from sklearn.metrics import r2_score
r2_score(y,predicted_data)  # as the correlation between input & output is exponential , accuracy is very low
```

```python
plt.scatter(x=dataset['Level'],y = dataset['Salary'])
plt.plot(dataset['Level'],predicted_data,color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=4)
X_tran = pf.fit_transform(X)
```

```python
lr = LinearRegression()
lr.fit(X_tran,y)
```

```python
pred = lr.predict(X_tran)
```

```python
r2_score(pred,y)
```

```python
plt.scatter(x=dataset['Level'],y = dataset['Salary'])
plt.plot(dataset['Level'],pred,color="red")
```
